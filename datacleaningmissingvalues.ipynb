# Further understand techniques on how to hanlde missing data
Missing values that python reads and recognizes as missing data: blank spaces etc


```python
#the provided code is using the pd.to_numeric function to convert the values into numeric values, even handling non convertible values like strings to NaN not a Number Values
df['NUM_BATH']=pd.to_numeric(df['NUM_BATH'],errors='coerce')
```

```python
df
```

Sometimes you might want to delete, other times you might want to replace
Method 1: replace all standard missing values in a certain column with one value

```python
#Replace missing values with a number
df['ST_NUM'].fillna(125,inplace=True)
```

```python
df
```

```python
#Location based replacement
df.loc[2,'ST_NUM']=130
```

```python
df
```

```python
#replacce using median
median=df['NUM_BEDROOMS'].median()
median
```

```python
df['NUM_BEDROOMS'].fillna(median,inplace=True)
```
